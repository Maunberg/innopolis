## Пропуски

```python
print(df.isna().sum())
```

Многие столбцы имеет пропуски, в целом обработка пропусков может быть индвидуальной и зависит от ситуации и задач, которые мы решаем.

В данном наборе данных предлагается:

1 Удалить игры без года создания (**year_of_release**) (пустых данных не так много, при этом для анализа такие данные непригодны, так как не дают полной картины для анализа).

2 Также можно удалить строки с пропущенными значениями в столбцах **name** и **genre** (всего 2 строки).

3 Большое количество пропусков в столбцах **rating**, **critic_score**, **user_score** - такие данные нельзя удалять из датасета.
При этом невозможно взять средние или медианные значения по столбцу, так как это не будет отражать реальную картину (можно в целом взять средние, но значения получатся не слишком "хорошими").
Для восстановления таких данных необходимо учесть слишком много информации - более логично сохранить строки и обозначить, что данные для этих игр не заполнены.

```python
df_games = df.copy() # создадим копию датасета и будем работать с ней
# (почему не сделали просто df_games = df: df2 = df1 не делает копию df1 и не присваивает это значение df2, а устанавливает указатель, который ссылается на df1. Таким образом, любые изменения в df2 приведут к изменениям в df1)
```

```python
df_games[df_games['year_of_release'].isna()] # игры в которых год не заполнен (Nan)
```

```python
df_games[df_games['name'].isna()] # игры где название не заполнено
```

```python
df_games[df_games['name'].isna()].index # можно удалить строки с данными индексами
```

```python
df_games.drop(df_games[df_games['year_of_release'].isna()].index,inplace=True) # убираем игры без года создания, так как они не нужны для аналитики, а восстановить год невозможно
df_games.drop(df_games[df_games['name'].isna()].index,inplace=True) # убираем игры без названий
```

```python
print(df_games.isna().sum())
```

Столбцы, в которых много пропущенных значений заменяем конкретными значениями

```python
df_games['rating'].fillna('не заполнен', inplace=True) # заполняем пропуски конкретным значением
df_games['critic_score'].fillna(0, inplace=True)
df_games['user_score'].fillna(0, inplace=True)
```

```python
print(df_games.isna().sum())
```

## Дубликаты

```python
print(df_games.duplicated().sum())
```

Явных дубликатов нет, удалять нечего.

Поищем скрытые дубликаты с помощью `unique`:

```python
col_for_dup = ['user_score', 'rating', 'platform', 'genre', 'name'] # нет смысла искать скрытые дубликаты во всех столбцах, обычно только в object и то не во всех
for col in col_for_dup:
  print('\n', col)
  print(df_games[col].unique())
```

Скрытых дубликатов не обнаружено. Однако, представляет интерес, что в столбце user_score есть значение tbd, а остальные данные числовые. По сути 'tbd' означает, что сейчас идет оценивание и по факту оценки нет, в целом все tbd можно заменить значением 0

```python
df_games.loc[df_games['user_score'] == 'tbd', 'user_score'] = 0 # "tbd" означает, что сейчас идет оценивание и по факту оценки нет, поэтому все "tbd" можно заменить
df_games['user_score'].value_counts()  # команда позволяет вывести количество каждого значения в датасете
```

```python
df_games['rating'].value_counts() # представляет интерес в целом рассмотреть как распределены значения в столбце
```

```python
df_games['critic_score'].value_counts()
```

В critic_score - данные целые числ, .0 не обязательно добавлять, это значит, что можно преобразовать к типу int

```python
df_games.info()
```

Пропущенных значений нет, но некоторые типы данных необходимо изменить.


*   год - целое число
*   critic_score - целое число
*   user_score - числовой столбец



```python
try:
    df_games['year_of_release'] = pd.to_numeric(df_games['year_of_release'], downcast='integer') # год это целое число
    df_games['critic_score'] = pd.to_numeric(df_games['critic_score'], downcast='integer') # оценка критиков это целые числа
    df_games['user_score'] = pd.to_numeric(df_games['user_score'], downcast='float', errors ='coerce') #  столбец типа float

except:
    print('Ошибка изменения типа данных')
```

Для работы с непредсказуемым поведением данных есть конструкция try-except. Принцип работы такой: исходный код помещают в блок try. Если при выполнении кода из блока try возникнет ошибка, воспроизведётся код из блока except.

Посмотрим сколько игр выпускалось для каждого жанра и создадим сводную таблицу `pt_games`

```python
# Построим сводную таблицу количество игр по жанрам
pt_games = pd.pivot_table(df_games, index='genre', values='name', aggfunc='count').reset_index()
pt_games.columns = ['genre','games_per_genre'] # переименуем название столбца с количеством игр
pt_games.sort_values(by = 'games_per_genre', ascending=False) # отсортируем по количеству игр по убыванию
```

*   для `bar` - по одной оси - *Жанр*, по другой оси - *Количество игр*
